**PRECEDENCE ASSOCIATIVITY:**
Precedence(ORDER OF EXECUTION) defines the priority of operators when evaluating an expression. Higher precedence operators are evaluated first.

PRECEDENCE RULE.PNG






ASSOSIATIVITY:
Associativity(DIRECTION OF EXECUTION) decides the direction in which operators with the same precedence are evaluated.

```python
# practice 1:
x = 5 + 3 * 2 ** 2 - (8 // 4) * 3
print(x)
#OUTPUT ??
# A/C precedence rule: 5+3*2**2-2*3=>5+3*4-2*3=>5+12-6=>17-6=>11

# practice 2:
x = 2 ** 3 ** 2
print(x)
# OUTPUT ??
# A/c to associativity rule: 2**9=>512


# Example 1: Operator Precedence
print(3 + 2 * 2)  # 3 + (2 * 2) = 7
print(5 ** 2 ** 2) # 5 ** (2 ** 2) = 5 ** 4 = 625

# Example 2: Left to Right Associativity
print(10 / 2 * 5)  # (10 / 2) * 5 = 5 * 5 = 25

# Example 3: Right to Left Associativity
print(2 ** 3 ** 2)  # 2 ** (3 ** 2) = 2 ** 9 = 512

# Example 4: Logical Operators
print(10 > 5 and 5 < 2)  # True and False = False
```

**Logical Operators: Used to combine conditional statements.**

Logical Operators:
***and***: Returns True only if both conditions are True.
 If one is False, the result is False.
***or***: Returns True if at least one condition is True.
 Only returns False if both are False.
***not***: Reverses the truth value.
 If the condition is True, it becomes False, and vice versa.

```python
x, y = True, False
print(x and y)  # False
print(x or y)  # True
print(not x)  # False

# Logical Operators:
# and: Returns True only if both conditions are True. If one is False, the result is False.
# or: Returns True if at least one condition is True. Only returns False if both are False.
# # not: Reverses the truth value. If the condition is True, it becomes False, and vice versa.
```

** Identity Operators**

Check if two variables reference the same object.

Identity Operators in Python (is and is not)

Identity operators compare the memory location of two objects to check if they are the same object (not just equal in value but actually the same in memory).

```python
# EXAMPLE
x = [1, 2, 3]  # x is assigned a list [1, 2, 3]
y = x          # y is assigned the same reference(MEMORY LOCATION)as x
print(x is y)  # True
print(x == y)  # True
```

identity-operators.PNG**Difference Between is and ==**

01: is(identity operator) checks if two variables refer to the same object in memory.

02: ==(equality operator) checks if the values of two objects are the same.

When Should You Avoid Using is?

❌ When comparing values (Use == instead).

❌ For mutable objects like lists, dictionaries, and sets, because Python creates new objects.

❌ For strings and numbers outside Python's caching range, as they may have different memory addresses.

Final Takeaway

✅ Use is for checking None.

✅ Use is when checking if two objects reference the same memory.

✅ Use is in singleton patterns (e.g., database connections).

✅ Use is not to ensure a variable is not None.


```python
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True (because the values are the same)
print(a is b)  # False (because they are different objects in memory)
# BUT
a = [1, 2, 3]  # False (Since order matters in lists, [1, 2, 3] is not the same as [3, 2, 1], so a == b returns False.)
b = [3, 2, 1]  # False (because they are different objects in memory)
print(a == b)  # Each time you create a list using [], Python allocates a new memory location for it.
print(a is b)  #Since a and b are stored at different locations, a is b returns False.


```
