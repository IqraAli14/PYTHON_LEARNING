# Introduction to Lists

List
Lists are used to store multiple items in a single variable.

Lists are one of 4 built-in data types in Python used to store collections of data.

Lists are created using square brackets:

```python
# Creating a List

# Lists are created using square brackets [].

my_list = ["red", "yellow", "green"]
print(my_list)
```

. List Characteristics

a. Ordered

Items have a defined order that does not change.

New items are added at the end of the list.

Some methods can change the order of items.

b. Changeable

Lists allow modification after creation.

Items can be added, removed, or updated.

c. Allow Duplicates

Lists can have duplicate values.

 lists are mutable, meaning you can modify their contents after creation. You can:

Add elements (append(), insert(), extend())

Modify elements (by indexing)

Remove elements (remove(), pop(), del)

Sort and reverse (sort(), reverse())

```python
# Example:

my_list = [1, 2, 3]
my_list.append(4)    # [1, 2, 3, 4]
my_list[1] = 10      # [1, 10, 3, 4]
my_list.pop(2)       # [1, 10, 4]
print(my_list)

# duplication in list:
thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
```

List Length

The len() function returns the number of items in a list.



```python
my_list = [10, 20, 30, 40, 50]
print(len(my_list))  # Output: 5

# The len() function counts all the elements, including duplicates and different data types.
my_list = [1, "hello", 3.14, True]
print(len(my_list))  # Output: 4

# List items can be of any data type.
```

In Python, lists are objects of the list data type. You can check this using the type() function.

Example:


my_list = [1, 2, 3, 4]
print(type(my_list))
  # Output: <class 'list'>
  
Since lists are objects, they come with built-in methods like .append(), .remove(), .sort(), etc

```python
my_list = [1, 2, 3, 4]
print(type(my_list))  # Output: <class 'list'>

my_list.append(5)
print(my_list)  # Output: [1, 2, 3, 4, 5]

my_list.remove(2)
print(my_list)  # Output: [1, 3, 4, 5]

my_list.sort()
print(my_list)  # Output: [1, 3, 4, 5]


my_list.reverse()
print(my_list)  # Output: [5, 4, 3, 1]

my_list.pop(2)
print(my_list)  # Output: [5, 4, 1]

my_list.clear()
print(my_list)  # Output: []

my_list.extend([6, 7, 8])
print(my_list)  # Output: [6, 7, 8]

my_list = my_list + [9, 10]
print(my_list)  # Output: [6, 7, 8, 9, 10]

my_list = ["apple", "banana", "cherry"]
print(type(my_list))  # Output: <class 'list'>
```

 Using list() Constructor

Lists can also be created using the list() constructor.

```python
my_list = list((1, 2, 3, 4, 5))  # Double parentheses needed for tuples
print(my_list)  # Output: [1, 2, 3, 4, 5]
```
