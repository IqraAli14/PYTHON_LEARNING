# Tuple Items
Tuple items are ordered, unchangeable, and allow duplicate values.
When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.



Tuple items are indexed, the first item has index [0], the second item has index [1] etc.

```python
tuple = (4, "ali", True, "ali", 4)
print(tuple[1]) #indexing
```

Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.

```python
# A constructor method is a special method used to initialize objects when they are created.
#  In Python, the __init__ method is the constructor method for classes.

#  Summary: Understanding __init__, self, and Direct Attribute Access in Python
# In Python, the __init__ method and self keyword are crucial for creating and managing objects in Object-Oriented Programming (OOP).
#  Here‚Äôs a breakdown of why they are necessary and how they work:

# 1Ô∏è‚É£ __init__ Method (Constructor)
# Purpose: The __init__ method is a special constructor that initializes object attributes automatically when an instance is created.
# Without __init__, we must manually assign attributes after creating an object, which is inefficient.
# Example:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person1 = Person("Ali", 25)  # Automatically sets attributes
print(person1.name)  # Output: Ali


# 2Ô∏è‚É£ self Keyword (Instance Reference)
# Purpose: self refers to the current instance of the class and ensures that attributes are unique to each object.
# Without self, attributes wouldn't belong to the object, leading to errors.

class Person:
    def __init__(self, name):
        self.name = name  # Assigns value to the object's attribute

person1 = Person("Ali")
person2 = Person("Ahmed")

print(person1.name)  # Output: Ali
print(person2.name)  # Output: Ahmed  (Separate data for each object)


# 3Ô∏è‚É£ Why Not Access Attributes Directly?
# If we don‚Äôt use __init__, we must manually set attributes, which is inefficient.
# Example (Without __init__):

class Person:
    pass

person1 = Person()
person1.name = "Ali"  # Manual assignment

# Problem: This approach lacks structure and is error-prone.
# ‚úÖ Using __init__ ensures proper initialization and automatic setup.


# 4Ô∏è‚É£ What Happens If We Remove self?
# If self is removed, Python doesn‚Äôt know which object‚Äôs attributes to assign.

# Incorrect Code (Without self):
class Person:
    def __init__(name):  # ‚ùå Missing 'self'
        name = name  # ‚ùå Does not assign to the object

person1 = Person("Ali")  # ‚ùå Error

# correct code (with sel)
class Person:
    def __init__(self, name):
        self.name = name  # ‚úÖ Assigns to instance

person1 = Person("Ali")
print(person1.name)  # ‚úÖ Output: Ali



# #
# 5Ô∏è‚É£ tuple() Constructor (Built-in Example)
# Python also has built-in constructors like tuple(), list(), and dict().
# Example

my_tuple = tuple([1, 2, 3])
print(my_tuple)  # Output: (1, 2, 3)


# #
# üîπ __init__ is essential for object initialization
# üîπ self ensures each object has separate data
# üîπ Avoid direct attribute access for better structure



```

```python
### **Python - Access Tuple Items**

#### **1Ô∏è‚É£ Access Tuple Items Using Index**
# - You can access tuple elements using their **index number** inside square brackets `[ ]`.
# - **Indexing starts from `0`** (first element).
# - Example:

thistuple = ("apple", "banana", "cherry")
print(thistuple[1])  # Output: banana


#### **2Ô∏è‚É£ Negative Indexing**
# - Negative indexing allows accessing elements from the **end** of the tuple.
# - `-1` refers to the **last item**, `-2` refers to the **second last item**, and so on.
# - Example:

thistuple = ("apple", "banana", "cherry")
print(thistuple[-1])  # Output: cherry


#### **3Ô∏è‚É£ Access a Range of Indexes (Tuple Slicing)**
# - You can specify a range using **`tuple[start:end]`**, where `start` is included and `end` is **excluded**.
# - **Example:** Extract items from index `2` to `4` (excluding index `5`).

thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])  # Output: ('cherry', 'orange', 'kiwi')


#### **4Ô∏è‚É£ Omitting Start or End Index**
# - **Start from the beginning up to a specific index:**

print(thistuple[:4])  # Output: ('apple', 'banana', 'cherry', 'orange')

# - **Start from a specific index to the end:**

print(thistuple[2:])  # Output: ('cherry', 'orange', 'kiwi', 'melon', 'mango')


#### **5Ô∏è‚É£ Range of Negative Indexes**
# - You can **specify negative indexes** to slice the tuple from the end.
# - **Example:** Extract items from index `-4` (included) to `-1` (excluded).

print(thistuple[-4:-1])  # Output: ('orange', 'kiwi', 'melon')

#### **6Ô∏è‚É£ Check if an Item Exists in a Tuple**
# - Use the **`in` keyword** to check if an item is present in the tuple.
# - Example:

thistuple = ("apple", "banana", "cherry")
if "apple" in thistuple:
      print("Yes, 'apple' is in the tuple")  # Output: Yes, 'apple' is in the tuple.

```

# python - update tuples

```python
### **Python - Update Tuples**

#### **1. Tuples are Immutable**
# - Tuples **cannot be changed, added to, or removed from** once created.
# - However, there are some workarounds to modify tuples indirectly.



#### **2. Changing Tuple Values**
# - Tuples cannot be modified directly, but a workaround exists:
  # 1. Convert the tuple into a **list**.
  # 2. Modify the list.
  # 3. Convert it back into a **tuple**.

# **Example:**
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)  # Output: ('apple', 'kiwi', 'cherry')



#### **3. Adding Items to a Tuple**
# Since tuples are immutable, they do not have a built-in `append()` method. However, there are two ways to add items:

##### **A. Convert Tuple into a List**
# - Convert the tuple into a **list**, add the item, and convert it back.

# **Example:**

thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)

##### **B. Concatenating Tuples**
# - You can add tuples together.
# - When adding a **single item**, remember to include a **comma** after it.

# **Example:**

thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y
print(thistuple)  # Output: ('apple', 'banana', 'cherry', 'orange')




#### **4. Removing Items from a Tuple**
# Tuples are immutable, so items **cannot be removed directly**. However, the workaround is:

##### **A. Convert Tuple into a List and Remove Item**
# **Example:**

thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)


##### **B. Delete the Entire Tuple**
# - Use `del` to completely delete a tuple.
# **Example:**

thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple)  # This will raise an error as the tuple no longer exists.

```

```python
# **Python - Unpack Tuples**

### Unpacking a Tuple
# - When creating a tuple, assigning values to it is called **packing**.
# - Extracting values back into variables is called **unpacking**.

#### Example: Packing a tuple

fruits = ("apple", "banana", "cherry")


#### Example: Unpacking a tuple

fruits = ("apple", "banana", "cherry")

green, yellow, red = fruits

print(green)  # Output: apple
print(yellow)  # Output: banana
print(red)  # Output: cherry

# - The number of variables must match the number of values in the tuple.
# - If not, use an **asterisk (`*`)** to collect remaining values into a list.

### Using Asterisk (`*`)
# - If there are more values than variables, adding `*` before a variable name stores extra values as a **list**.

#### Example: Assigning extra values to a list

fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

green, yellow, *red = fruits

print(green)  # Output: apple
print(yellow)  # Output: banana
print(red)  # Output: ['cherry', 'strawberry', 'raspberry']


# - If `*` is used on a variable other than the last one, values are assigned accordingly until the remaining values match the number of remaining variables.

#### Example: Using `*` in a different position

fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

green, *tropic, red = fruits

print(green)  # Output: apple
print(tropic)  # Output: ['mango', 'papaya', 'pineapple']
print(red)  # Output: cherry


```
